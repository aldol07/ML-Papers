```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(x,0)

x = np.linspace(-10,10,100)
y = relu(x)

plt.plot(x,y)
plt.title('ReLU Activation Function')
```

```python
def leaky_relu(x):
    return np.maximum(0.1*x,x)

x = np.linspace(-10,10,100) 
y = leaky_relu(x)

plt.plot(x,y)
plt.title('Leaky ReLU Activation Function')
```

```python
def prelu(x,alpha):
    return np.maximum(alpha*x,x)

x = np.linspace(-10,10,100)
y = prelu(x,0.5)

plt.title('PReLU Activation Function (alpha=0.5)')
plt.plot(x,y)
```

```python
def elu(x,alpha):
  return np.where(x >= 0,x,alpha*(np.exp(x)-1))

x=np.linspace(-10,10,100)
y=elu(x,0.2)
plt.title("eLU")
plt.plot(x,y)
```
